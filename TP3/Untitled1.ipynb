```python
def sam (limit,size):
    lista = []
    for i in range(size):
        poly = []
        for j in range(n):
            poly.append(randint(1,limit))
        lista.append(Rq(poly))
    
    res = matrix(Rq,size,1,lista)
    return res

def Decompose(C,alfa):

    r = mod(C,int(q))
    r0 = int(mod(r, int(alfa)))

    if (r-r0) == (q-1):
        r1 = 0
        r0 = r0 - 1
    else:
        r1 = (r-r0)/(int(alfa))

    return (r1,r0)

def auxHB(r):
    
    print(r)
    res = Decompose(r,2*gama2)
        
    return res[0]

def HighBits(polys):
    
    lista = polys.list()
    
    for i in range(len(lista)):
        poly = lista[i]
        polyL = poly.list()
        
        for j in range(len(polyL)):
            polyL[j] = auxHB(int(polyL[j]))
        
        lista[i] = polyL

    return lista
```
